#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO
normalized_prices = {}

# Normalizar precios por población
for pop in south_belt:
    prices = ds_south_belt[ds_south_belt["level5"] == pop]["price"]
    norm = (prices - prices.min()) / (prices.max() - prices.min())
    normalized_prices[pop] = norm

# Graficar los histogramas en el mismo plot
plt.figure(figsize=(10,6))
plt.hist(
    [normalized_prices[pop] for pop in south_belt],
    bins=10,
    label=south_belt,
    alpha=0.6,
    edgecolor="black"
)

plt.title("Normalized Price Distribution - South Belt of Madrid")
plt.xlabel("Normalized Price")
plt.ylabel("Frequency")
plt.legend(title="Population")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps, CircleMarker, LayersControl

# Map centered on south west part of Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
## HERE: plot the coordinates of the estates

# Crear diccionario con coordenadas por población
coords_dict = {}

for pop in ds["level5"].unique():
    subset = ds[ds["level5"] == pop]
    coords_dict[pop] = list(zip(subset["latitude"], subset["longitude"]))


## PUT HERE YOUR CODE:
colors = {
    "Fuenlabrada": "red",
    "Leganés": "blue",
    "Getafe": "green",
    "Alcorcón": "orange"
}

# Agregar marcadores al mapa
for pop, coords in coords_dict.items():
    for lat, lon in coords:
        marker = CircleMarker()
        marker.location = (lat, lon)
        marker.radius = 5
        marker.color = colors.get(pop, "black")
        marker.fill_color = colors.get(pop, "black")
        marker.fill_opacity = 0.7
        map.add_layer(marker)

# Agregar control de capas
map.add_control(LayersControl())
map
```
